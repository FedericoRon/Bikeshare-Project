```python
import pandas as pd

df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks H-Farm/raw_meteo_data/meteo_data.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks H-Farm/bike_data/bike_data.csv')
```

```python
df1.drop('Unnamed: 0', axis = 1, inplace= True)
df2.drop('Unnamed: 0', axis = 1, inplace= True)
```

```python
df2['Date time'] = pd.to_datetime(df2['started_at']).dt.floor('h')
df2
```

```python
df2['Date time'] = df2['Date time'].dt.strftime('%m/%d/%Y %H:%M:%S')
df2
```

```python
#df = pd.concat([df1, df2], axis = 1, keys=[df1['Date time']])
df = pd.merge(df2, df1, on ='Date time')
df
```

```python
df.drop(columns= ['Name'], inplace= True)
```

```python
df.columns
```

```python
df =  df[['rideable_type', 'Date time', 'started_at', 'ended_at',
       'start_station_name', 'start_station_id', 'end_station_name',
       'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng',
       'member_casual', 'Maximum Temperature',
       'Minimum Temperature', 'Temperature', 'Wind Chill', 'Heat Index',
       'Precipitation', 'Snow', 'Snow Depth', 'Wind Speed', 'Wind Direction',
       'Wind Gust', 'Visibility', 'Cloud Cover', 'Relative Humidity',
       'Conditions']]
df
```

```python
df.sort_values(by=['Date time'], inplace= True)
df
```

```python
df['ended_at'] = pd.to_datetime(df['ended_at'])
df['started_at'] = pd.to_datetime(df['started_at'])
```

```python
df['Duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds()
df
```
